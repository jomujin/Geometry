```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import random as rd
from shapely.geometry import *
```

```python
# 삼각형의 점 만들기(랜덤범위1, 랜덤범위2, 점 개수)
# 삼각형의 경우 3점의 위치만 달라도 문제없는 형태 구성이 가능함
def make_pt(r1, r2, n):
    x_list, y_list = [], []
    while len(x_list) < n and len(y_list) < n:
        x, y = rd.randint(r1, r2), rd.randint(r1, r2)
        if x not in x_list and y not in y_list:
            x_list.append(x)
            y_list.append(y)
    return [(x, y) for x, y in zip(x_list, y_list)]
```

```python
# 삼각형의 무게중심
def centroid_tri(pts):
    sum_x, sum_y = 0, 0
    for pt in pts:
        sum_x += pt[0]
        sum_y += pt[1]
    centroid = [sum_x/3, sum_y/3]
    return centroid
```

```python
fig, ax = plt.subplots(1, 4, figsize=(18, 4))
    
for i in range(len(ax)):
    pts = make_pt(0, 10, 3)
    centroid = centroid_tri(pts)
    tri = Polygon(pts)
    c_pt = Point(centroid)
    ax[i].plot(*tri.exterior.xy)
    ax[i].scatter(c_pt.x, c_pt.y)
    ax[i].set_xlim(-1, 11)
    ax[i].set_ylim(-1, 11)
plt.show()
```
